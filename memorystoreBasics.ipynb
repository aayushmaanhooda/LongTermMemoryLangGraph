Memory stores


```python
from dotenv import load_dotenv
from langgraph.store.memory import InMemoryStore


load_dotenv()
```

```python
# create a store
store = InMemoryStore()

# creating a namespace
namespace = ("user", "u1")
```

Creating a memory

```python
# adding memory
store.put(namespace , "1", {"data": "user likes pizza"})
store.put(namespace , "2", {"data": "user prefers dark mode"})
```

```python
#  another namespace
namespace2 = ("user", "u2")

# adding memory
# inside put we haev the actual namespce, the unique id for that memory , and the actual memory
store.put(namespace2 , "1", {"data": "user likes pasta"})
store.put(namespace2 , "2", {"data": "user prefers grid style navigation"})
```

Retrieving Memories

```python
#  store.get(namespace, key)
store.get(namespace, "1")
store.get(namespace2, "1")
```

Retrieving all memories

```python
items = store.search(namespace)

for item in items:
    print(item.value.get("data"))
```

Semantic search

```python
from langchain_openai import OpenAIEmbeddings

embedding_model = OpenAIEmbeddings(model="text-embedding-3-small")

store = InMemoryStore(index={"embed": embedding_model , "dims": 1536})
```

```python
namespace = ("users",)
```

```python
store.put(namespace , "1", {"data": "user likes example in python"})
store.put(namespace , "2", {"data": "user gets up eraly in moring and hit gym daily"})
store.put(namespace , "3", {"data": "user is breathing langchain recently"})
store.put(namespace , "4", {"data": "user likes pasta nowdays"})
store.put(namespace , "5", {"data": "user prefers grid style navigation"})
store.put(namespace , "6", {"data": "user is based in sydney"})
```

```python
items = store.search(namespace , query="what topic is user leanring now a days?", limit =1)

# items
for item in items:
    print(item.value.get("data"))
```
