PART 1 MEMORY STORE

```python
from dotenv import load_dotenv
from langgraph.store.memory import InMemoryStore

load_dotenv()
```

```python
# creating a store
# this is inhertied from BaseStore and gives you get, search, put , delete and ist async version as well
store = InMemoryStore()
```

```python
# creating a namspace , we create it as tuple, thatsd how docs says
# think of it as users isone folder inside it there is other folder called u1
namespace = ("users", "u1")
```

```python
# creating memories
# inside put fucntion we have the actual namespace, the unique id for that memory , and the actual memory
# namespace, key , value
store.put(namespace , "1", {"data": "User likes F1"})
store.put(namespace , "2", {"data": "User prefers grid style navigation"})
```

```python
# retrieving a memories

store.get(namespace, "1")
```

```python
items = store.search(namespace)

for item in items:
    print(item.value["data"])
```
